```python
import openai
import pandas as pd

# Load your data
data = pd.read_csv('challenge2_data.csv')  # Assume your data is in a CSV file named data.csv

# Filter dfs via fms, using the fm_list as keys for those df
fm_list = ['LEAKAGE',
           'DRAG FORCE',
           'VIBRATION',
           'NOISE',
           'AESTHETICAL DEFECT',
           'ELECTRICAL SYSTEM FAULT',
           'SPONGY/NOT BLEEDABLE',
           'CORROSION',
           'BREAKAGE/DEFORMATION',
           'WEAR/ABRASION',
           'DEGRADATION OF PISTON OR CALIPER BODY STROKE']

df_dict ={}
for item in fm_list:
  df_dict[item] = data[data['fm'] == item]

# Loop through each list and use language model to explain why the fm makes sense, going row by row. Take that explanation and make a document for each fm
import openai
import pandas as pd

# Assume fm_list and df_dict are already defined
# ...

key = input('openai key')
openai.api_key = key

f=[]
r=[]

for fm in fm_list:
    df = df_dict[fm]
    
    for index, row in df.iterrows():
        if index < 10:
            messages = [
                {
                    'role': 'system',
                    'content': '''rewrite the claim without mentioning customer or 'claim'. Do not include the failure mode.  Restrict the response to 500 characters '''
                },
                {
                    'role': 'user',
                    'content': f'''{row['claim']}
    failure mode: {row['fm']}'''
                }
            ]
            response = openai.ChatCompletion.create(
                model="gpt-3.5-turbo",  # Assuming you're using the base GPT-3.5 model
                messages=messages,
                max_tokens=150,  # Adjusted to allow for more detailed responses
                temperature=0.5,  # A moderate temperature value for a balance of creativity and coherence
            )

            explanation = response['choices'][0].message
            document_content = f"Row {index}:\nWarranty Claim: {row['claim']}\nFailure Mode: {row['fm']}\nExplanation: {explanation}\n\n"
            print(document_content)
            f.append(row['fm'])
            r.append(response['choices'][0]['message']['content'])
print(f)
print(r)
            


    # Write the document content to a file, one file per failure mode
    # file_path = f'/content/drive/MyDrive/Brembo/{fm}_explanation.txt'
    # with open(file_path, 'w') as file:
    #   file.write(document_content)


data = {'claim': r, 'fm': f}
df = pd.DataFrame(data)

df
```
